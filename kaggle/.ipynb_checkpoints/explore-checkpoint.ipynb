```python
import cv2, numpy as np
img = cv2.imread("in.jpg", cv2.IMREAD_UNCHANGED)
up = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_LANCZOS4)  # or INTER_CUBIC
# Unsharp mask
blur = cv2.GaussianBlur(up, (0,0), 1.0)
sharp = cv2.addWeighted(up, 1.5, blur, -0.5, 0)
cv2.imwrite("out.jpg", sharp)
```
